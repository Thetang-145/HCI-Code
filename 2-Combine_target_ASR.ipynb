```python
import pandas as pd
```

```python
participants = []
for i in range(16):
    if (i+1)<10:
        participants.append("Participant0"+str(i+1))
    else:
        participants.append("Participant"+str(i+1))
```

```python
df = pd.read_csv("Dataset\Result - Total.csv", index_col=0)
```

```python
df = df[df.index.notnull()]
df.head()
```

```python
cond_table = pd.read_csv("Dataset\Cond_table.csv", index_col=0)
cond_table.head()
```

```python
df_target = pd.read_csv("Dataset\Target-4-16.csv")
df_target[18:]
```

```python
col = ["Participant", "Corpus", "Condition", "Trail", "Target_Sentence", "ASR_Sentence"]
df_combine = pd.DataFrame(columns=col)

Corpus_list = list((df.index).unique())
for p in participants:
    df_p = df[p]
    # Select different target corpus
    if p == "Participant1":
        df_target = pd.read_csv("Dataset\Target-1.csv")
    elif p == "Participant2" or p == "Participant3":
        df_target = pd.read_csv("Dataset\Target-2-3.csv")
    else:
        df_target = pd.read_csv("Dataset\Target-4-16.csv")
        
    for c in Corpus_list:
        df_p_c = df_p[df_p.index == c].reset_index(drop=True)
        df_target_c = df_target[df_target.corpus == c].reset_index(drop=True)
        n = len(df_p_c)
        for i in range(n):
            cond = ((cond_table.loc[[p],[c]]).values)[0][0]
            target =  df_target_c.texts[i]
            ASR = df_p_c[i]
            row = pd.DataFrame([[p, c, cond, i+1, target, ASR]], columns=col)
            df_combine = pd.concat([df_combine, row])

df_combine
```

```python
df_combine.to_csv("Dataset/Result_Transform.csv", index=None)
```
