* WER: Word Error Rate
* MER: Match Error Rate
* WIL: Word Information Lost

```python
ground_truth = "Hey Google did the Liverpool win yesterday"
hypothesis = "3 hey Google yesterday"

ground_truth = "Hey Google turn on"
hypothesis = "Hey Google turn off 4:50 a.m"



ground_truth = "Hey Google set the thermostat to 20 degrees"
hypothesis = "Hey Google set up 20 to here"

transformation = jiwer.Compose([
    jiwer.ToLowerCase(),
    jiwer.RemoveWhiteSpace(replace_by_space=True),
    jiwer.RemoveMultipleSpaces(),
    jiwer.RemoveSpecificWords(["?"]),
    jiwer.ReduceToListOfListOfWords(word_delimiter=" ")
]) 

measures = jiwer.compute_measures(
    ground_truth, 
    hypothesis,
    truth_transform=transformation, 
    hypothesis_transform=transformation
)
print(measures['wer'])
print(measures['mer'])
print(measures['wil'])
```
